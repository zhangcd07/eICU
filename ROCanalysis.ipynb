```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('backbone.csv')
```

```python
AMS=pd.read_csv('AMS.csv')
heartrate110=pd.read_csv('heartrate110.csv')
historyofcancer=pd.read_csv('historyofcancer.csv')
historyofchroniclung=pd.read_csv('historyofchroniclung.csv')
historyofheartfailure=pd.read_csv('historyofheartfailure.csv')
RR30=pd.read_csv('RR30.csv')
sao290=pd.read_csv('sao290.csv')
systolic100=pd.read_csv('systolic100.csv')
temperature36=pd.read_csv('temperature36.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
for i in df.index:
    df.loc[i, 'AMS']=df.loc[i, 'patientunitstayid'] in list(AMS['patientunitstayid'])
for i in df.index:
    df.loc[i, 'heartrate110']=df.loc[i, 'patientunitstayid'] in list(heartrate110['patientunitstayid'])
for i in df.index:
    df.loc[i, 'historyofcancer']=df.loc[i, 'patientunitstayid'] in list(historyofcancer['patientunitstayid'])
for i in df.index:
    df.loc[i, 'historyofchroniclung']=df.loc[i, 'patientunitstayid'] in list(historyofchroniclung['patientunitstayid'])
for i in df.index:
    df.loc[i, 'historyofheartfailure']=df.loc[i, 'patientunitstayid'] in list(historyofheartfailure['patientunitstayid'])
for i in df.index:
    df.loc[i, 'RR30']=df.loc[i, 'patientunitstayid'] in list(RR30['patientunitstayid'])
for i in df.index:
    df.loc[i, 'sao290']=df.loc[i, 'patientunitstayid'] in list(sao290['patientunitstayid'])
for i in df.index:
    df.loc[i, 'systolic100']=df.loc[i, 'patientunitstayid'] in list(systolic100['patientunitstayid'])
for i in df.index:
    df.loc[i, 'temperature36']=df.loc[i, 'patientunitstayid'] in list(temperature36['patientunitstayid'])
```

```python
df['morbidobesity']=df['bmi']>40
```

```python
df
```

```python
df['ismale']=df['gender']=='Male'
```

```python
df
```

```python
df['PESI']=df['age']+10*df['ismale']+30*df['historyofcancer']+10*df['historyofheartfailure']+10*df['historyofchroniclung']+20*df['heartrate110']+30*df['systolic100']+20*df['RR30']+20*df['temperature36']+60*df['AMS']+20*df['sao290']
```

```python
df
```

```python
df['obPESI']=df['age']+10*df['ismale']+30*df['historyofcancer']+10*df['historyofheartfailure']+10*df['historyofchroniclung']+20*df['heartrate110']+30*df['systolic100']+20*df['RR30']+20*df['temperature36']+60*df['AMS']+20*df['sao290']-20*df['morbidobesity']
```

```python
df
```

```python
from sklearn.metrics import roc_curve, auc
```

```python
df['death']=df['actualhospitalmortality']=='EXPIRED'
```

```python
fpr_PESI, tpr_PESI, threshold_PESI = roc_curve(df['death'], df['PESI'])
aucPESI=auc(fpr_PESI,tpr_PESI)
aucPESI
```

```python
fpr_obPESI, tpr_obPESI, threshold_obPESI = roc_curve(df['death'], df['obPESI'])
aucobPESI=auc(fpr_obPESI,tpr_obPESI)
aucobPESI
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
lw = 2
plt.figure(figsize=(10,10))
plt.plot(fpr_PESI, tpr_PESI, color='darkorange', 
         lw=lw, label='ROC curve conventional PESI (area = %0.3f)' % aucPESI) 
plt.plot(fpr_obPESI, tpr_obPESI, color='darkgreen',
         lw=lw, label='ROC curve obesity PESI (area = %0.3f)' % aucobPESI)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic curve')
plt.legend(loc="lower right")
plt.show()
```
